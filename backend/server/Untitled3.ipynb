```python
import os
from PIL import Image
import numpy as np

# Define the directories
input_dir = '/Volumes/Chris_SSD/file_shuttle/cells'
output_dir = '/Volumes/Chris_SSD/file_shuttle/cells2'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Function to crop the image to the relevant area
def crop_image(image):
    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Find the bounding box of the non-zero regions
    non_zero_coords = np.argwhere(image_array > 0)
    top_left = non_zero_coords.min(axis=0)
    bottom_right = non_zero_coords.max(axis=0) + 1  # +1 to include the bottom right pixel

    # Crop the image using the bounding box coordinates
    cropped_image_array = image_array[top_left[0]:bottom_right[0], top_left[1]:bottom_right[1]]
    cropped_image = Image.fromarray(cropped_image_array)

    return cropped_image

# Iterate through the images in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.tif') or filename.endswith('.tiff'):
        # Open the image
        image_path = os.path.join(input_dir, filename)
        image = Image.open(image_path)

        # Crop the image
        cropped_image = crop_image(image)

        # Save the cropped image to the output directory
        output_path = os.path.join(output_dir, filename)
        cropped_image.save(output_path)

        print(f'Cropped and saved: {filename}')

print('All images have been processed.')
```
