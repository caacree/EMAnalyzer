```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec, GridSpecFromSubplotSpec

colors = {
    'nucleus':      (0.85, 0.25, 0.30),  # muted coral red
    'ER':           (0.00, 0.80, 0.40),  # bright teal to really stand out
    'Glycogen':     (1.00, 0.85, 0.00),  # a punchy, vibrant yellow
    'Mitochondria': (0.60, 0.30, 0.85),  # soft violet
}

# assume em, nucleus, er_2, glycogen_2, mito_2 are already defined as 2D arrays

fig = plt.figure(figsize=(15, 8))

# outer grid: 2 rows × 2 cols, left column twice as wide as right (2/3 vs 1/3)
outer = GridSpec(2, 2, figure=fig, width_ratios=[2, 1], wspace=0.2, hspace=0.2)

# inner grid in the left column: 2 rows × 3 cols
inner = GridSpecFromSubplotSpec(2, 3, subplot_spec=outer[:, 0], wspace=0.1, hspace=0.1)
inner_axes = [fig.add_subplot(inner[i, j]) for i in range(2) for j in range(3)]

# display five panels + one blank slot
images = [em, nucleus, er_2, glycogen_2, mito_2]
titles = ['EM', 'Nucleus', 'ER', 'Glycogen', 'Mitochondria']

for ax, img, title in zip(inner_axes, images, titles):
    ax.imshow(img, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

# leave the last (6th) subplot blank
inner_axes[-1].axis('off')

# composite overlay in the right column spanning both rows
ax_comp = fig.add_subplot(outer[:, 1])
ax_comp.imshow(em, cmap='gray', alpha=0.3)

# more subdued, “earth‐tone” colors
colors = {
    'nucleus':      (0.7, 0.2, 0.2),   # muted brick red
    'ER':           (0.2, 0.5, 0.2),   # olive green
    'Glycogen':     (0.2, 0.2, 0.6),   # deep indigo
    'Mitochondria': (0.5, 0.4, 0.1),   # golden brown
}

mask_dict = {
    'nucleus':      (nucleus > 0),
    'ER':           (er_2       > 0),
    'Glycogen':     (glycogen_2 > 0),
    'Mitochondria': (mito_2     > 0),
}

for name, mask in mask_dict.items():
    # create an H×W×4 RGBA overlay
    overlay = np.zeros(mask.shape + (4,), dtype=float)
    overlay[mask, :3] = colors[name]
    overlay[mask,  3] = 1.0
    ax_comp.imshow(overlay, interpolation='none')

ax_comp.set_title('All Structures (no LD)')
ax_comp.axis('off')

plt.show()
```
